```python
import sys
import os
sys.path.append(os.path.abspath(".."))
import spyh
```

```python
points = spyh.ParticleInitialize(gap = 0.1)
points.rectangle(3,3)
physics = spyh.PhysicsAttributes(VISCOSITY=20, SOUND_SPEED=1000)

particle = spyh.ParticleAttributes(0.1*1.5, 'bspline')
particle._initialize_particle(points, physics)

sim_config = spyh.SimAttributes(bound=[3,3],collision_coefficient=0.99, Dt=0.001, substep=10, totaltime=0.300)
sim = spyh.Simulation(sim_config, particle)
sim.update_fulltime('leapfrog')
anim = sim.create_anim()
```

```python
sim.save_anim(anim,'water_drop')
```
